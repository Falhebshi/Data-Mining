##Goal

The purpose of this project is to analyze and predict the productivity of garment factory employees
using data mining techniques. Our main goal is to classify whether employees can achieve their
target productivity or not, and to cluster employees based on similar productivity patterns.

- *Type of ML Task*: Classification and Clustering  
- *Domain/Problem*: Workforce productivity in the garment manufacturing industry. This problem
is important because productivity plays a key role in the efficiency and success of the industry,
and predicting productivity can help managers optimize scheduling, incentives, and workforce
allocation.


## Dataset Source

Dataset : https://www.kaggle.com/datasets/ishadss/productivity-prediction-of-garment-employees
